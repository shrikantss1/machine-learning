# Using RFE (Recursive Feature Elimination)

```python
#Step1: Initialize ML Algo
from sklearn.linear_model import LinearRegression
modelAlgo = LinearRegression()
```

```python
#Step2: Initialize RFE
#step refers to how many feature cols to eliminate per iteration

from sklearn.feature_selection import RFE
selectFeatures = RFE(estimator=modelAlgo, step = 1)
```

```python
selectFeatures.fit(features, labels)
```

```python
selectFeatures.ranking_
```

```python
#Guideline: Select the columns with ranking 1 and 2
#RFE selected 'California', 'Florida', 'New York', 'R&D Spend'
```

# Select From Model


```python
#Step1: Initialize ML algo
from sklearn.linear_model import LinearRegression
modelAlgo = LinearRegression()
```

```python
#Step2: Initialize SFM

from sklearn.feature_selection import SelectFromModel

selectFeatures = SelectFromModel(estimator=modelAlgo)
```

```python
#Step3: Fit the data to the SFM
selectFeatures.fit(features, labels)
```
