```python
features = data.iloc[:, [0,1]].values
labels = data.iloc[:, [2]].values
```

```python
import warnings
warnings.filterwarnings("ignore")
SL = 0.2
CL = 1 - SL

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

max_test_score = 0
optimal_rs = 0

for rs in range(1,300):
    X_train, X_test, y_train, y_test = train_test_split(
        features,
        labels,
        test_size=0.2,
        random_state=rs
    )

    model = KNeighborsClassifier()
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)

    if test_score > train_score and test_score > CL :
        if test_score > max_test_score:
            max_test_score = test_score
            optimal_rs = rs
        print(f"Test Score: {test_score}, Train Score: {train_score}, RS: {rs}")


print(f"Best Test Score: {max_test_score}, Optimal : {optimal_rs}")
```

# **KNN Regression Implementation**

```python
data = pd.read_csv("../../data/50_Startups.csv")
```

```python
features = data.iloc[:, [0,1,2,3]].values
labels = data.iloc[:, [4]].values
```

```python
from sklearn.preprocessing import OneHotEncoder
oheState = OneHotEncoder(sparse_output=False)
stateDummy = oheState.fit_transform(features[:, [3]])
```

```python
finalFeatureSet = np.concatenate((stateDummy, features[:, [0,1,2]]), axis= 1)
```

```python
finalFeatureSet
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

CL = 0.8

max_test_score = - np.inf
best_rs = None
best_k = None

for k in range(3,11):
    for i in range(1,100):
        X_train,X_test, y_train,y_test = train_test_split(finalFeatureSet, labels, test_size=0.2, random_state=i)
        model = KNeighborsRegressor(n_neighbors=k)

        model.fit(X_train, y_train)
        test_score = model.score(X_test, y_test)
        train_score = model.score(X_train, y_train)

        if test_score > train_score and test_score > CL and test_score > max_test_score:
            max_test_score = test_score
            best_rs = i
            best_k = k



print(f"Max TestScore: {max_test_score} k: {best_k}, RandonState: {best_rs}")
```
