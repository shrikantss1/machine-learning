```python
# Binary classification use - case
# We are dealing with UnBalanced Dataset
```

```python
features = data.iloc[:,[0,1]].values
labels = data.iloc[:, [2]].values
```

```python
#
import warnings
warnings.filterwarnings("ignore")
SL = 0.2

CL = 1 - SL


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

max_test_score = 0
optimal_rs = 0

for rs in range(1,300):
    X_train, X_test, y_train, y_test = train_test_split(
        features,
        labels,
        test_size=0.2,
        random_state=rs
    )

    model = LogisticRegression()
    model.fit(X_train, y_train)

    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)

   
    if test_score > train_score and test_score >= CL:
        if test_score > max_test_score:
            max_test_score = test_score
            optimal_rs = rs
        print(f"TestScore is {test_score}, Train Score is {train_score}, rs is {rs}")

print("-----------------------")
print(f"Max test score: {max_test_score}, Optimal Ransom Sample: {optimal_rs}")
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    features,
    labels,
    test_size=0.2,
    random_state=220
)

model = LogisticRegression()
model.fit(X_train, y_train)

train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)


if test_score > train_score and test_score >= CL:
    print(f"TestScore is {test_score}, Train Score is {train_score}")
```

```python
#classification report

from sklearn.metrics import classification_report

print(classification_report(labels, model.predict(features)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(labels, model.predict(features))
```
