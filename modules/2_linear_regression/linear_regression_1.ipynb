```python
#Goal: Create a model that can predict the salary of the employee based on
#employee's years of experience
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../data/Salary_Data.csv")
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
features = data.iloc[:,[0]].values
labels = data.iloc[:, [1]].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

CL = 0.95

max_test_score = 0.0;
optimal_seed = 0;
for seed in range(1,100):
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=seed)
    model = LinearRegression()
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
   
    if test_score > train_score and test_score >= CL:
        if test_score > max_test_score:
            max_test_score = test_score
            optimal_seed = seed
        print(f"Test Score: {test_score} Train Score: {train_score} RandomState: {seed}")

print(f"---- Max Test Score {max_test_score} Optimal Seed: {optimal_seed}----")
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=30)

model = LinearRegression()
model.fit(X_train, y_train)

```

```python
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print(f" Test Score: {test_score} Train Score: {train_score}")
```

```python
yExperience = 10 #float(input("Enter years of Experience: "))

salary = model.predict(np.array([[yExperience]]))

print(f"The Predicted salary for {yExperience} years of experience is {salary[0][0]:0.2f}")
```
