Seperate my data as features and label

```python
features = data.iloc[:,[0,1,2]].values
labels = data.iloc[:,[3]].values
```

```python
features
```

```python
labels
```

```python
#Imputation
#scikitlearn package

#Step1: Import the relevant package
from sklearn.impute import SimpleImputer

# Imputer class for Country
siForCountry = SimpleImputer(missing_values=np.nan, strategy="most_frequent")

# Fit the object with the data column
siForCountry.fit(features[:, [0]])
features[:,[0]] = siForCountry.transform(features[:, [0]])
```

```python
siForCountry.statistics_
```

```python
features
```

```python
# Age Column

siForAge = SimpleImputer(missing_values=np.nan, strategy="median")
siForAge.fit(features[:,[1]])

features[:, [1]] = siForAge.transform(features[:, [1]])
```

```python
features
```

```python
# Salary

siForSalary = SimpleImputer(missing_values=np.nan, strategy="mean")

features[:, [2]] = siForSalary.fit_transform(features[:,[2]])
```

```python
features
```

```python
#Categorical Data Handling
#Recommended to create Dummy Variables
#In Sklearn you can achieve the same using OneHotEncoder

from sklearn.preprocessing import OneHotEncoder

oheContry = OneHotEncoder(sparse_output=False)
fCountry = oheContry.fit_transform(features[:, 0].reshape(-1, 1))

fCountry
```

```python
featuresFinalSet = np.concatenate((fCountry, features[:,[1,2]]), axis=1)
```
