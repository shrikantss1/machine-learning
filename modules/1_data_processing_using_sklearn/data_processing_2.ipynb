```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../../data/melb_data.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
featurestat=data.iloc[:, [12,14,15,15]].values
featureDomain=data.iloc[:,[12,14,15,16]].values
```

```python
from sklearn.impute import SimpleImputer
```

```python
siForCar = SimpleImputer(missing_values=np.nan, strategy="mean")
```

```python
siForCar.fit(featurestat[:, [0]])
```

```python
featurestat[:, [0]] = siForCar.transform(featurestat[:,[0]])
```

```python
siCarDomain = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=1)
siCarDomain.fit(featureDomain[:, [0]])
siCarDomain.transform(featureDomain[:,[0]])
```

```python
print(f"Car Stat Calculated: {siForCar.statistics_}")
print(f"Car Domain Calculated: {siCarDomain.statistics_}")
```

```python
siArea = SimpleImputer(missing_values=np.nan, strategy="mean")
featurestat[:,[1]] = siArea.fit_transform(featurestat[:, [1]])

siAreaDomain = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=80)
featureDomain[:,[1]] = siArea.fit_transform(featureDomain[:, [1]])
```

```python
siYear = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
featurestat[:, [2]] = siArea.fit_transform(featurestat[:, [2]])

siYearDomain = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=200)
featureDomain[:, [2]] = siYearDomain.fit_transform(featureDomain[:, [2]])
```
