```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv("../../data/Social_Network_Ads.csv")
```

```python
data.info()
```

```python
features = data.iloc[:,[0,1]].values
label = data.iloc[:,2].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
algo = DecisionTreeClassifier()
```

```python
algo.fit(X_train, y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(features, label, clf=algo, legend=2)
```

```python
pred_dt = algo.predict(X_test)
accuracy_score(y_test, pred_dt)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=82)
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(features,label,clf=model,legend=2)
```

```python
pred_logistic_regression = model.predict(X_test)
accuracy_score(y_test, pred_logistic_regression)
```

```python
from sklearn.svm import SVC

model = SVC(kernel="rbf")
model.fit(X_train, y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(features,label,clf=model,legend=2)
```

```python
pred_svc = model.predict(X_test)
accuracy_score(y_test, pred_svc)
```
